# Spotify Listening Time — Exploratory Data Analysis
Objectif : prédire `listening_time` via régression linéaire.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
df = pd.read_csv('../data/raw/spotify_churn_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Distribution de la target : `listening_time`

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df['listening_time'].hist(bins=40, ax=axes[0])
axes[0].set_title('Distribution listening_time')
df.boxplot(column='listening_time', ax=axes[1])
axes[1].set_title('Boxplot listening_time')
plt.tight_layout()
plt.show()
```

## Corrélations avec la target

```python
numerical_cols = ['age', 'songs_played_per_day', 'skip_rate', 'ads_listened_per_week', 'offline_listening', 'listening_time']
corr = df[numerical_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Matrice de corrélation')
plt.tight_layout()
plt.show()
```

## Features catégorielles vs listening_time

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
for ax, col in zip(axes.flatten(), ['subscription_type', 'device_type', 'gender', 'country']):
    df.groupby(col)['listening_time'].mean().sort_values().plot(kind='bar', ax=ax)
    ax.set_title(f'listening_time moyen par {col}')
    ax.set_ylabel('listening_time')
    ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```
